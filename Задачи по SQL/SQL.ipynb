```python
query = '''
SELECT 
    AVG(ratings.rating) AS rating,
    COUNT(distinct reviews.review_id) AS review_id,
    books.title AS title 
FROM 
    ratings
INNER JOIN reviews ON reviews.book_id = ratings.book_id 
INNER JOIN books ON reviews.book_id = books.book_id 
GROUP BY
title
ORDER BY 
review_id DESC'''
pd.io.sql.read_sql(query,con = engine)
```

- Определите издательство, которое издало наибольшее число книг толще 50 страниц (так вы исключите из анализа различные брошюры);

```python
query = '''
SELECT 
    COUNT(books.title) AS title, 
    publishers.publisher  AS publisher
FROM 
    books
INNER JOIN 
    publishers ON publishers.publisher_id  = books.publisher_id 
WHERE 
    books.num_pages IN
    (SELECT
    books.num_pages 
    FROM 
        books
    WHERE 
        books.num_pages>50)
GROUP BY 
publisher
ORDER BY 
title DESC
LIMIT 1

'''
pd.io.sql.read_sql(query,con = engine)
```

- Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более пользовательскими оценками;

```python
query = '''
SELECT 
    books.book_id AS book_id,
    COUNT(ratings.rating_id)AS rating_id,
    AVG(ratings.rating) AS rating,
    authors.author AS author
    
FROM 
    books
INNER JOIN  
    ratings ON ratings.book_id = books.book_id
INNER JOIN 
    authors ON authors.author_id = books.author_id
WHERE ratings.rating IN
(SELECT 
   MAX(ratings.rating) AS  rating
FROM 
    ratings
GROUP BY
rating)
    
GROUP BY
    books.book_id,
   
    author
    
HAVING 
    COUNT(rating_id) >= 50
ORDER BY 
    rating DESC
LIMIT 1
'''
pd.io.sql.read_sql(query,con = engine)
```

- Посчитайте среднее количество текстовых обзоров пользователей, которые поставили более чем по 50 оценок.

```python
query = '''
SELECT 
    
    AVG(Sub.count_review) AS avg_count_review
    
FROM
    (SELECT 
        COUNT(distinct reviews.review_id) as count_review,
        reviews.username AS user_name,
        COUNT(ratings.rating_id) AS rating_id
    FROM 
        reviews
    LEFT JOIN ratings ON ratings.username = reviews.username
    
    GROUP BY 
        user_name 
        
    HAVING
        COUNT(rating_id) > 1000
) AS Sub


'''
pd.io.sql.read_sql(query,con = engine)
```
